```python
def plot_per_class_distributions(feature):
    df0 = df[df['Class'] == 0]
    df1 = df[df['Class'] == 1]

    sns.histplot(
        df0[feature], bins=50, color='r', kde=True, 
        stat='density', linewidth=0, label='NoFraud')
    
    sns.histplot(
        df1[feature], bins=50, color='b', kde=True, 
        stat='density', linewidth=0, label='Fraud')
    
    plt.title(feature)
    plt.legend()
    plt.show()
```

```python
def print_metrics(x, y, dataset_label):
    # Assuming you have a model already trained, for example:
    model = GaussianNB()
    model.fit(x_train, y_train)
    
    # Make predictions
    y_pred = model.predict(x)
    y_prob = model.predict_proba(x)[:, 1]
    
    # Calculate metrics
    accuracy = accuracy_score(y, y_pred)
    conf_matrix = confusion_matrix(y, y_pred)
    class_report = classification_report(y, y_pred)
    roc_auc = roc_auc_score(y, y_prob)
    f1 = f1_score(y, y_pred)

    # Print the metrics
    print(f"Metrics for {dataset_label} dataset:")
    print(f"Accuracy: {accuracy}")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Classification Report:")
    print(class_report)
    print(f"AUC: {roc_auc}")
    print(f"F1 Score: {f1}")
    print("\n")

# Now you can call this function with your training data
print_metrics(x_train, y_train, 'train')
```
