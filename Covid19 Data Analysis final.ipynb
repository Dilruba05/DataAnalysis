Find Maximum Infection rate for China, Italy and Spain

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

Find Maximum Ifection rate for all of the countries 

```python
countries = list(corona_dataset_aggregated.index)
max_infections_rates = []
for c in countries: 
    max_infections_rates.append(corona_dataset_aggregated.loc[c].diff().max())
    
max_infections_rates
```

```python
corona_dataset_aggregated["max_infections_rates"] = max_infections_rates
```

```python
corona_dataset_aggregated.head()
```

Craete a Dataframe ( New) with only nedded column

```python
coronadata= pd. DataFrame(corona_dataset_aggregated["max_infections_rates"])
```
